```python
# a histogram with matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# decide the names on the x-axis
x_names = ['t5-small', 'bert', 't5-base', 't5-large',  'rquge']
# set the y-axis name
plt.ylabel('Number of parameters')
# set the x-axis name
plt.xlabel('Model')
# decide the values on the y-axis
y_values = [60, 110, 220, 770,  900]

# decide the colors
colors = ['red', 'orange', 'green', 'darkblue',  'gray']

# plot the histogram
plt.bar(x_names, y_values, color=colors)


# plt.yscale('log')
# save the plot
plt.savefig('histogram.eps')
```
