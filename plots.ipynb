```python
# a histogram with matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.figure(figsize=(10, 5))
# decide the names on the x-axis
x_names = ['t5-tiny','t5-small', 'distilbert', 'bert', 'mt5-base', 't5-base', 't5-large',  'rquge', 'GPT-3.5']
# set the y-axis name
plt.ylabel('Number of parameters')
# set the x-axis name
plt.xlabel('Model')
# decide the values on the y-axis
y_values = [16e6, 60e6, 66e6, 110e6, 220e6, 220e6, 770e6,  900e6, 175000e6]

# decide the colors
# colors = ['cyan','red', 'y', 'darkorange', 'green', 'darkblue',  'gray', 'purple']

# decide the y-axis ticks
plt.yticks(np.arange(0, 1200, 100))

# start from 0 the basis of the y axis



colors = plt.get_cmap('viridis_r')
rescale = lambda y: (y - np.min(y))/(np.max(y) - np.min(y))

# plot the histogram
plt.ylim(1e6, 1e12)

plt.bar(x_names, y_values, color=colors(rescale((np.log10(y_values)))))

plt.yscale('log')

# save the plot
plt.savefig('histogram.eps')
```
